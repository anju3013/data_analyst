4. Fill in all the missing values present in all the columns in the dataset


```python
data['Age']=data['Age'].fillna(data['Age'].mean())
data['Cabin']=data['Cabin'].fillna(data['Cabin'].mode()[0])
data['Embarked']=data['Embarked'].fillna(data['Embarked'].mode()[0])
```

5. Check and handle outliers in at least 3 columns in the dataset


```python
graphs = data.select_dtypes(include = ['float64','int64'])
graphs.hist(figsize = (20,10))
plt.show()
```

```python
num_columns=['Age','Fare', 'SibSp']
for i in num_columns:
  plt.figure()
  plt.boxplot(data[i])
  plt.title([i])
  plt.show()
```

    Outlier handling for the column Age

```python
num_columns=['Age','Fare', 'SibSp']
for i in num_columns:
  print("outlier handling for column",i)
  Q1 = data[i].quantile(0.25)
  Q2 = data[i].quantile(0.50)
  Q3 = data[i].quantile(0.75)

  print("Q1 is: ",Q1)
  print("Q2 is: ",Q2)
  print("Q2 is: ",Q2)

  IQR = Q3-Q1
  print(IQR)

  lower_lim = Q1 - 1.5*IQR
  upper_lim = Q3 + 1.5*IQR

  print("lower whisker: ",lower_lim)
  print("upper whisker: ",upper_lim)

  outlier =[]
  for x in data[i]:
    if((upper_lim)):
      outlier.append(x)

  print('outliers for ',i,outlier)

  data[i] = data[i].clip(lower=lower_lim, upper=upper_lim)

  outlier_1 =[]
  for x in data[i]:
    if((upper_lim)):
      outlier.append(x)

  print('outliers after handling:',outlier_1)

  plt.boxplot(data['Age'])
  plt.show()
```
