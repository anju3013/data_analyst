1. The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.

```python
#2 Sample t Test
```

Here, we want to determine any increase in sales after stepping into digital marketing. Here the sample size is less than 30 and no information is given about the population standard deviation.So we are performing a 2 sample t test

H0: There is no difference in sales after stepping into digital marketing.

HA:There is an increase in sales after stepping into digital marketing

We choose level of significance as 0.05 as a criterion for hypothesis testing.


```python
from scipy.stats import ttest_ind
```

```python
t_stat,p_value=ttest_ind(df['Sales_before_digital_add(in $)'],df['Sales_After_digital_add(in $)'])
```

```python
p_value
```

```python
if p_value<0.05:
  print("Null hypothesis rejected")
else:
  print("Failed to reject null hypothesis")
```

As the null hypothesis is rejected as p_value <0.05(the level of significance).
Thus we can conclude that there is an increase in sale after stepping into digital marketing.

```python
#chi squre test of independence
```

Here we need to check there is any dependeny between the features 'Region' and 'Manager'.
Here we are performing chi square test for independence between these two.
Ho: There is no interdependency between 'Region' and 'Manager'
HA:There is some interdependency between 'Region' and 'Manager'



```python
c_table=pd.crosstab(df['Manager'],df['Region'])
c_table
```

```python
from scipy.stats import chi2_contingency
```

```python
chi2_stat,p_value1,dof,exp=chi2_contingency(c_table)
p_value1
```

```python
if p_value1<0.05:
  print("Null hypothesis rejected")
else:
  print("Failed to reject null hypothesis")
```

As per chi squre test we failed to prove that there is some interdependency between 'Region' and 'Manager'.
