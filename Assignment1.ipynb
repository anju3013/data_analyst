```python
# SL histogram
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(data['SL'],color='green')
plt.xlabel('SL')
plt.title('Histogram of SL')
plt.show()
```

```python
#Bar chart of Classification vs PW
plt.bar(data['Classification'],data['PW'])
plt.xlabel('Model')
plt.ylabel('PW')
plt.title('Classification Vs PW')
plt.show()
```

```python
#Scatter plot of PL and PW
plt.scatter(data['PL'],data['PW'],marker='o',color='purple')
plt.xlabel('PL')
plt.ylabel('PW')
plt.title('PL Vs PW')
plt.grid()
plt.show()
```
